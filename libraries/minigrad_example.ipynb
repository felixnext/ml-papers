```python
from mini_grad import Parameter, Value
import matplotlib.pyplot as plt
%matplotlib inline

 # generate a simple example
a = Value(2.0, name="a")
b = Value(3.0, name="b")
c = a * b; c.name = "c"

res = c.forward()
print(res)

bb = c.backward()
print(bb)

# plot graph
c.plot()
```

```python
from mini_grad import Neuron, Parameter
from mini_grad.ops import Sigmoid, Tanh
from random import random

num_ins = 4
ins = [Parameter(random(), name=f"x{i}") for i in range(num_ins)]
n1 = Neuron(ins, activation=Sigmoid, name="neuron_1")
n2 = Neuron(ins, activation=Tanh, name="neuron_2")

res = n1 + n2

res.plot(max_level=0)
```

```python
res.backward()
res.plot()
```
